## Changer la taille d'une image

```python
#On affiche la taille de l'image
img = mpimg.imread('PokemonData/Bulbasaur/001.png')
taille = img.shape
print(f'Taille: ', taille)
plt.imshow(img)
```

```python
#On change la taille
nouvelle_taille = (100, 100)
img_petite = cv2.resize(img, nouvelle_taille)
taille = img_petite.shape
print(f'Taille: ', taille)
plt.imshow(img_petite)
```
